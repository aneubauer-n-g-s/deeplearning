```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
%ll
```

```python
image = tf.image.decode_jpeg( tf.read_file("lagotto-romagnolo.jpg"))
```

```python
init_op = tf.global_variables_initializer() 
sess = tf.InteractiveSession() 
sess.run(init_op)
img = sess.run(image)
```

```python
print(img.shape)
```

```python
fig = plt.figure()
plt.imshow(img) 
plt.show()
```
